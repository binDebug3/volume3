```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    crimes = pd.read_csv("crime_data.csv", index_col="Year")
    crimes.drop(["Population", "Total"], axis=1).plot(title="Crime Rates in the US", figsize=(10, 10))
    
    crimes.drop(["Population", "Total"], axis=1).plot(subplots=True, layout=(3,3), sharey=True, 
                                                      title="Crime Rates in the US", figsize=(10, 10),
                                                      )
    crimes.plot(kind='hist',y=['Burglary','Larceny','Property'], 
                alpha=.7,bins=10,title="Frequency of Burglary, Larceny, and Property Crimes per Year")

    
prob1()
```

# Problem 2

```python
def build(crimes, xlabel, ylabel):
    crimes.plot(kind="scatter", x=xlabel, y=ylabel, 
                alpha=.8,xlim=(0,max(crimes[xlabel])+1), 
                ylim=(0,max(crimes[ylabel])+1), 
                title=f"{xlabel} vs {ylabel}")

def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    crimes = pd.read_csv("crime_data.csv", index_col="Year")

    interests = ["Violent", "Burglary", "Aggravated Assault"]
    for interest in interests:
        build(crimes, "Forcible Rape", interest)

    return (True, False, True)

prob2()   
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    crimes = pd.read_csv("crime_data.csv", index_col="Year")

    crimes.plot(kind="box", y=["Burglary","Violent","Vehicle Theft"], title="Distribution of Burglary, Violent, and Vehicle Theft")

    crimes.plot(kind="hexbin", x="Vehicle Theft", y="Robbery", gridsize=20, sharex=False, title="Distribution of Vehicle Theft against the values of Robbery")


prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    colleges = pd.read_csv("college.csv")

    build(colleges, "Expend", "Grad.Rate")
    plt.title("How does Expenditure affect Graduation Rate?")
    plt.xlabel("Expenditure in dollars")
    plt.ylabel("Graduation Rate in percent")

    colleges.plot(kind="hexbin", x="Room.Board", y="Personal", gridsize=20, sharex=False, title="Distribution of Room and Board against Personal Expenses")
    plt.xlabel("Room and Board Expenses in dollars")
    plt.ylabel("Personal Expenses in dollars")

    colleges.plot(kind="box", y=["Top10perc","Top25perc"], title="Percentage of Students from the Top of their HS Class")
    # change the x labels for each box plot
    plt.xticks([1,2], ["Top 10%", "Top 25%"])
    plt.ylabel("Percentage of Students")


prob4()
```
